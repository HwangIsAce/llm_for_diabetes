```python
import re

def count_tokens_approximate(text):
    if not isinstance(text, str):
        raise ValueError("Input must be a string.")
    
    text = text.lower()
    
    text = re.sub(r"\s+", " ", text.strip())
    
    tokens = re.findall(r"\w+|[^\s\w]", text, re.UNICODE)
    
    byte_tokens = sum(len(token.encode("utf-8")) for token in tokens)
    
    return len(tokens)

print(f"텍스트 토큰 개수: {count_tokens_approximate(result)}")
```

```python
import pandas as pd
import json

# Step 1: model_output 컬럼에서 아침, 점심, 저녁 값 추출
def extract_meals(model_output):
    """model_output JSON 문자열에서 아침, 점심, 저녁 값을 추출"""
    try:
        meals = json.loads(model_output) if isinstance(model_output, str) else {}
        if isinstance(meals, dict):
            return meals.get("Breakfast", ""), meals.get("Lunch", ""), meals.get("Dinner", "")
        return "", "", ""
    except json.JSONDecodeError:
        return "", "", ""

# Step 1-1: Breakfast, Lunch, Dinner 컬럼 생성
if 'Breakfast' not in output_daily.columns:
    output_daily[['Breakfast', 'Lunch', 'Dinner']] = output_daily['model_output'].apply(
        lambda x: pd.Series(extract_meals(x))
    )

# Step 2: dfh의 title에서 각 식사 값을 찾아 ingredients와 Total Carbohydrates 추출
def find_ingredients_and_carbs(title, dfh):
    """
    dfh에서 title에 해당하는 ingredients와 Total Carbohydrates 값을 반환.
    """
    match = dfh[dfh['title'] == title]
    if not match.empty:
        ingredients = match['ingredients'].iloc[0]
        nutrition_facts = match['nutrition_facts'].iloc[0]
        try:
            # nutrition_facts에서 Total Carbohydrates 추출
            nutrition_facts_dict = json.loads(nutrition_facts.replace("'", '"'))
            total_carbs = nutrition_facts_dict.get('Amount per Serving', {}).get('Total Carbohydrates', {}).get('Amount', 'N/A')
        except json.JSONDecodeError:
            total_carbs = 'N/A'
        return ingredients, total_carbs
    return None, 'N/A'

# Step 3: ingredients_with_carbs 컬럼 생성
output_daily['ingredients_with_carbs'] = output_daily.apply(
    lambda row: {
        "Breakfast": find_ingredients_and_carbs(row['Breakfast'], dfh),
        "Lunch": find_ingredients_and_carbs(row['Lunch'], dfh),
        "Dinner": find_ingredients_and_carbs(row['Dinner'], dfh)
    },
    axis=1
)

# 결과 확인
# print(output_daily[['Breakfast', 'Lunch', 'Dinner', 'ingredients_with_carbs']].head())
```

```python
import pandas as pd
import json

# Step 1: ingredients 컬럼에서 label과 탄수화물 질량 추출
def parse_and_extract_labels_with_carbs(ingredients):
    """
    ingredients 컬럼에서 각 식사 시간별로 label 값과 탄수화물 질량(carbohydrates)을 추출.
    """
    try:
        result = {}
        # 주어진 값이 dict인지 확인
        if isinstance(ingredients, dict):
            for meal_time, meal_data in ingredients.items():
                items, carbs = meal_data  # ('JSON 문자열', '탄수화물 질량')
                if isinstance(items, str):
                    # JSON 문자열을 파싱
                    parsed_items = json.loads(items.replace("'", '"'))
                    # 각 아이템의 'label' 값 추출
                    labels = [item.get('label', '') for item in parsed_items if 'label' in item]
                    result[meal_time] = {"labels": labels, "carbohydrates": carbs}
                else:
                    result[meal_time] = {"labels": [], "carbohydrates": "N/A"}
        return result
    except (json.JSONDecodeError, TypeError) as e:
        # 파싱 실패 시 빈 딕셔너리 반환
        return {"Breakfast": {"labels": [], "carbohydrates": "N/A"},
                "Lunch": {"labels": [], "carbohydrates": "N/A"},
                "Dinner": {"labels": [], "carbohydrates": "N/A"}}

# Step 2: ingredients 컬럼 확인 및 새로운 컬럼 생성
output_daily['ingredients_v2'] = output_daily['ingredients_with_carbs'].apply(parse_and_extract_labels_with_carbs)

# 결과 확인
# print(output_daily[['ingredients', 'ingredients_v2']].head())

```

```python
import openai
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()
# OpenAI API Key 설정
openai.api_key = os.getenv("OPENAI_API_KEY")

# 참고 GI 정보
reference_gi = result

# Step 1: GL 계산 함수 정의
def calculate_gl(labels, carbs, meal_type):
    """
    GPT API를 사용하여 GL 계산.
    labels: 식재료 리스트
    carbs: 전체 탄수화물 양
    meal_type: Breakfast, Lunch, Dinner 중 하나
    """
    try:
        # Prompt 생성
        prompt = f"""
        Calculate the Glycemic Load (GL) for the following {meal_type} ingredients based on their Glycemic Index (GI) and total carbohydrates ({carbs} grams).

        The formula for calculating GL is:
        GL = (GI × Carbohydrate content in grams) / 100.

        Instructions:
        1. Since only the total carbohydrate content is provided, evenly distribute the total carbohydrate content across all the listed ingredients.
        2. Use the provided Reference GI list: {reference_gi}.
        3. If an ingredient is not in the Reference GI list, do the following:
           a. Select the most similar ingredient from the Reference GI list.
           b. If no similar ingredient is found, use your knowledge of the ingredient's GI to estimate the GL.
           c. If calculating GL is still not possible, exclude the ingredient from the calculation.

        Ingredients: {', '.join(labels)}

        Respond with the total GL value for the meal, assuming the carbohydrates are evenly distributed, and briefly mention any assumptions or exclusions made during the calculation.
        """

        # GPT API 호출
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0125",
            messages=[
                {"role": "system", "content": "You are an expert in nutrition."},
                {"role": "user", "content": prompt},
            ],
            temperature=0.2,
            max_tokens=100,
        )

        # API 결과 추출
        gl_value = response['choices'][0]['message']['content'].strip()
        return gl_value

    except Exception as e:
        print(f"Error calculating GL for {meal_type}: {e}")
        return None

# Step 2: 각 row에 대해 GL 계산 및 데이터프레임 업데이트
def process_gl_for_row(row):
    """
    각 row의 Breakfast, Lunch, Dinner에 대해 GL 계산.
    """
    gl_results = {}

    for meal_time in ['Breakfast', 'Lunch', 'Dinner']:
        meal_data = row['ingredients_v2'].get(meal_time, {})
        labels = meal_data.get('labels', [])
        carbs = meal_data.get('carbohydrates', 'N/A')
        
        # carbs가 'N/A'가 아닌 경우에만 GL 계산
        if labels and carbs != 'N/A':
            gl_results[meal_time] = calculate_gl(labels, carbs, meal_type=meal_time)
        else:
            gl_results[meal_time] = None

    return gl_results

# Step 3: 데이터프레임의 각 row에 대해 GL 계산 적용
output_daily['GL'] = output_daily.apply(process_gl_for_row, axis=1)

# Step 4: None 값을 제외하고 GL 평균 계산
def calculate_gl_average(gl_results):
    """
    GL 결과에서 None 값을 제외한 평균 계산.
    """
    gl_values = []
    for meal_time in gl_results:
        try:
            # GL 값을 숫자로 변환
            gl_value = float(re.search(r"[-+]?\d*\.\d+|\d+", str(gl_results[meal_time])).group())
            gl_values.append(gl_value)
        except (ValueError, AttributeError):
            continue  # None 또는 파싱 실패 시 건너뜀

    if gl_values:
        return sum(gl_values) / len(gl_values)
    return None

# Step 5: 각 row의 GL 평균 계산
output_daily['GL_Average'] = output_daily['GL'].apply(calculate_gl_average)

# 결과 확인
print(output_daily[['GL', 'GL_Average']].head())
```

```python
import pandas as pd
import json
import re

# Step 1: 추천된 식사 이름 추출 함수
def extract_recommended_meal(model_output):
    """
    model_output에서 추천된 식사 이름을 추출합니다.
    """
    try:
        match = re.search(r"^(.*?) is recommended", model_output)
        if match:
            return match.group(1).strip()
        return None
    except Exception:
        return None

# Step 2: dfh의 title에서 일치하는 ingredients와 carbohydrates 값을 찾는 함수
def find_ingredients_and_carbs(title, dfh):
    """
    dfh에서 title에 해당하는 ingredients 값과 Total Carbohydrates의 Amount 값을 반환합니다.
    """
    match = dfh[dfh['title'] == title]
    if not match.empty:
        ingredients = match['ingredients'].iloc[0]
        nutrition_facts = match['nutrition_facts'].iloc[0]
        try:
            # JSON 변환 시 문자열 포맷 문제 해결
            formatted_nutrition_facts = nutrition_facts.replace("'", '"').replace("None", "null")
            nutrition_data = json.loads(formatted_nutrition_facts)
            
            # Total Carbohydrates 값 추출
            carbs = nutrition_data.get('Amount per Serving', {}).get('Total Carbohydrates', {}).get('Amount', 'N/A')
        except (json.JSONDecodeError, TypeError) as e:
            # 에러 발생 시 문제 있는 데이터를 출력
            print(f"Error parsing nutrition_facts for title '{title}': {nutrition_facts}")
            carbs = 'N/A'
        return ingredients, carbs
    return None, 'N/A'

# Step 3: output_alt의 새 컬럼 recommended_meal 생성
output_alt['recommended_meal'] = output_alt['model_output'].apply(extract_recommended_meal)

# Step 4: recommended_meal을 기반으로 ingredients와 carbohydrates 값을 매칭하여 새 컬럼 생성
output_alt['ingredients_and_carbs'] = output_alt['recommended_meal'].apply(
    lambda meal: find_ingredients_and_carbs(meal, dfh)
)

# 결과 확인
# print(output_alt[['model_output', 'recommended_meal', 'ingredients', 'total_carbohydrates']].head())

```

```python
import pandas as pd
import json

# Step 1: ingredients 컬럼에서 label 값과 carb 값 추출
def parse_and_extract_labels_and_carbs(ingredients, carbs):
    """
    ingredients 컬럼에서 label 값과 carb 값을 추출.
    """
    try:
        # ingredients 처리: 문자열인지 확인 후 JSON 배열로 변환
        if isinstance(ingredients, str):
            ingredients = json.loads(ingredients.replace("'", '"'))  # 문자열을 JSON으로 변환

        # label 값 추출
        labels = [item.get('label', '') for item in ingredients if isinstance(item, dict) and 'label' in item] if isinstance(ingredients, list) else []

        # carb 값 처리: 숫자로 변환 가능 여부 확인
        carb_value = carbs if isinstance(carbs, str) else 'N/A'
        return {'labels': labels, 'carb': carb_value}

    except (json.JSONDecodeError, TypeError):
        # JSON 변환 실패 시 기본값 반환
        return {'labels': [], 'carb': 'N/A'}

# Step 2: ingredients_v2 컬럼 생성
output_alt['ingredients_v2'] = output_alt.apply(
    lambda row: parse_and_extract_labels_and_carbs(row['ingredients_and_carbs'][0], row['ingredients_and_carbs'][1]), axis=1
)

# 결과 확인
# print(output_alt[['ingredients', 'total_carbohydrates', 'ingredients_v2']].head())
```

```python
import openai
import pandas as pd
from dotenv import load_dotenv
import os
import re

# 환경 변수에서 API 키 로드
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

# 참고 GI 정보
reference_gi = result

# Step 1: GL 계산 함수 정의
def calculate_gl(labels, carbs):
    """
    GPT API를 사용하여 GL 계산.
    labels: 식재료 리스트
    carbs: 전체 탄수화물 양 (예: '25g')
    """
    try:
        if not labels or carbs == 'N/A':
            return None  # 입력 값이 없는 경우 None 반환

        # carbs에서 'g' 제거 및 float 변환
        carbs_value = float(carbs.replace('g', '').strip())

        # Prompt 생성
        prompt = f"""
        Calculate the Glycemic Load (GL) for the following ingredients based on their Glycemic Index (GI) and total carbohydrates ({carbs_value} grams).

        The formula for calculating GL is:
        GL = (GI × Carbohydrate content in grams) / 100.

        Instructions:
        1. Since only the total carbohydrate content is provided, evenly distribute the total carbohydrate content across all the listed ingredients.
        2. Use the provided Reference GI list: {reference_gi}.
        3. If an ingredient is not in the Reference GI list:
           a. Select the most similar ingredient from the Reference GI list.
           b. If no similar ingredient is found, use your knowledge of the ingredient's GI to estimate the GL.
           c. If calculating GL is still not possible, exclude the ingredient from the calculation.
        4. Respond with only the total GL value as a numeric value (e.g., 15.6). Do not include any additional text or explanation.

        Ingredients: {', '.join(labels)}
        """

        # GPT API 호출
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0125",
            messages=[
                {"role": "system", "content": "You are an expert in nutrition."},
                {"role": "user", "content": prompt},
            ],
            temperature=0.2,
            max_tokens=50,
        )

        # API 결과 추출
        gpt_response = response['choices'][0]['message']['content'].strip()

        # 응답에서 숫자만 추출
        match = re.search(r"[-+]?\d*\.\d+|\d+", gpt_response)  # 숫자값 추출
        if match:
            return float(match.group())  # 숫자값 반환
        else:
            print(f"Invalid response: {gpt_response}")
            return None

    except Exception as e:
        print(f"Error calculating GL: {e}")
        return None

# Step 2: 각 row에 대해 GL 계산 및 데이터프레임 업데이트
def process_gl_for_row(row):
    """
    각 row의 ingredients_v2에서 GL 계산.
    """
    meal_data = row['ingredients_v2']
    labels = meal_data.get('labels', [])
    carbs = meal_data.get('carb', 'N/A')
    
    # GL 계산
    if labels and carbs != 'N/A':
        return calculate_gl(labels, carbs)
    return None

# Step 3: 데이터프레임의 각 row에 대해 GL 계산 적용
output_alt = output_alt.copy()  # SettingWithCopyWarning 방지
output_alt['GL_Response'] = output_alt.apply(process_gl_for_row, axis=1)  # GPT 응답 원문 저장
output_alt['GL'] = output_alt['GL_Response']

# Step 4: None 값을 제외한 GL 평균 계산
def calculate_average_gl(gl_list):
    """
    GL 리스트에서 None 값을 제외하고 평균 계산.
    """
    try:
        gl_values = [float(gl) for gl in gl_list if gl is not None]
        return sum(gl_values) / len(gl_values) if gl_values else None
    except ValueError:
        return None

# GL 평균 계산
output_alt['GL_Average'] = output_alt['GL']

# 결과 확인
print(output_alt[['ingredients_v2', 'GL_Response', 'GL']].head())
```
