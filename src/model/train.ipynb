```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token # Must add EOS_TOKEN
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # Must add EOS_TOKEN, otherwise your generation will go on forever!
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("passionMan/diabetes_v4", split = "train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported
from datasets import Dataset, concatenate_datasets
import pandas as pd
import numpy as np

# ë¦¬ìŠ¤íŠ¸ í˜•íƒœì˜ datasetì„ Pandas DataFrameìœ¼ë¡œ ë³€í™˜
dataset_df = pd.DataFrame(dataset)

# ê¸´ ë°ì´í„°ì™€ ì§§ì€ ë°ì´í„° ë¶„ë¦¬
long_data = dataset_df[dataset_df['dataset'].isin(['pubmedqa', 'icliniq', 'pubmed'])].reset_index(drop=True)  # ì¸ë±ìŠ¤ ì œê±°
short_data = dataset_df[~dataset_df['dataset'].isin(['pubmedqa', 'icliniq', 'pubmed'])].reset_index(drop=True)  # ì¸ë±ìŠ¤ ì œê±°

# ë°ì´í„°ì…‹ ìƒì„±
long_dataset = Dataset.from_pandas(long_data)
short_dataset = Dataset.from_pandas(short_data)

# ë°ì´í„° ê²°í•©
combined_dataset = concatenate_datasets([short_dataset, long_dataset])

# ìƒ˜í”Œë§ ê°€ì¤‘ì¹˜ ì„¤ì •
# Hugging Face Datasetì—ì„œ map í•¨ìˆ˜ê°€ dictë¥¼ ë°˜í™˜í•˜ë„ë¡ ìˆ˜ì •
def compute_weights(example):
    return {"weight": 0.1 if example['dataset'] in ['pubmedqa', 'icliniq', 'pubmed'] else 0.9}

# map í•¨ìˆ˜ë¡œ ê°€ì¤‘ì¹˜ ë¦¬ìŠ¤íŠ¸ ìƒì„±
weighted_dataset = combined_dataset.map(compute_weights)

# ê°€ì¤‘ì¹˜ ì¶”ì¶œ ë° ì •ê·œí™”
weights = np.array(weighted_dataset["weight"])
weights = weights / weights.sum()  # ê°€ì¤‘ì¹˜ í•©ì´ 1ì´ ë˜ë„ë¡ ì •ê·œí™”

# ë°ì´í„° ìž¬ìƒ˜í”Œë§
weighted_indices = np.random.choice(len(combined_dataset), size=len(combined_dataset), p=weights)
resampled_dataset = combined_dataset.select(weighted_indices)

# í•™ìŠµ ì„¤ì •
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=resampled_dataset,
    dataset_text_field="text",
    max_seq_length=2048,  # ê¸´ ë°ì´í„°ë„ ì²˜ë¦¬ ê°€ëŠ¥í•˜ë„ë¡ ì„¤ì •
    dataset_num_proc = 16,  # í˜„ìž¬ 2ì—ì„œ ì¦ê°€

    args=TrainingArguments(
        per_device_train_batch_size=32,
        gradient_accumulation_steps=1,
        warmup_steps=50,
        num_train_epochs=2,  # 2 epochs ì„¤ì •
        learning_rate=3e-5,  # ë‚®ì€ í•™ìŠµë¥ ë¡œ ê· í˜• ìœ ì§€
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        optim="adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        logging_steps=50,
        seed=3407,
        save_steps=200,
        output_dir="outputs/weighted_sampling2",
        report_to="none",
    ),
)

```

```python
import random
import numpy as np
import pandas as pd  # pandas ìž„í¬íŠ¸ ì¶”ê°€
from datasets import load_dataset, Dataset, DatasetDict

# ì‹œë“œ ê°’ ì„¤ì •
SEED = 42

# Pythonì˜ random ëª¨ë“ˆ ì‹œë“œ ê³ ì •
random.seed(SEED)

# NumPy ì‹œë“œ ê³ ì •
np.random.seed(SEED)

# 1. ë°ì´í„°ì…‹ ë¡œë“œ
dataset_name = "passionMan/diabetes_v4"
train_dataset = load_dataset(dataset_name, split="train")
test_dataset = load_dataset(dataset_name, split="test")

# 2. ì¤‘ë³µ í™•ì¸ì„ ìœ„í•œ train ë°ì´í„°ì˜ (input, output) ì„¸íŠ¸ ìƒì„±
train_pairs = set((row["input"], row["output"]) for row in train_dataset)

# 3. ê° taskë³„ ìƒ˜í”Œ ê°œìˆ˜ ì„¤ì •
task_sample_limits = {
    'qa_objective_medqa': 50,
    'qa_objective_medmcqa': 50,
    'alternative_diet': 50,
    'daily_diets': 50,
    'ie_extract_relation': 50,
    'nli': 50,
    "summarization": 50,
    "qa_descriptive": 50,
    "qa_subjective": 50,
    "generation": 50,
}
default_sample_limit = 100  # ë‚˜ë¨¸ì§€ taskëŠ” 100ê°œì”© ìƒ˜í”Œë§

# 4. taskë³„ ìƒ˜í”Œë§ ë°ì´í„° ì´ˆê¸°í™”
task_sampled_data = {}
# task_countsë¥¼ task_sample_limitsë¡œ ì´ˆê¸°í™”
task_counts = {task: 0 for task in task_sample_limits}

# 5. Taskë³„ ìƒ˜í”Œë§
# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì„ ì…”í”Œë§í•˜ì—¬ ëžœë¤ ìƒ˜í”Œë§ ë³´ìž¥
test_dataset = test_dataset.shuffle(seed=SEED)

for row in test_dataset:
    task = row["task"]
    dataset_value = row.get("dataset")  # dataset ì»¬ëŸ¼ ê°’ í™•ì¸
    input_output_pair = (row["input"], row["output"])

    # qa_objective ì„¸ë¶„í™”
    if task == "qa_objective":
        if dataset_value == "medqa":
            task = "qa_objective_medqa"  # medqaë¡œ ì„¸ë¶„í™”
        elif dataset_value == "medmcqa":
            task = "qa_objective_medmcqa"  # medmcqaë¡œ ì„¸ë¶„í™”
        else:
            continue  # medqa, medmcqaê°€ ì•„ë‹ˆë©´ ê±´ë„ˆë›°ê¸°

    # í•´ë‹¹ taskì˜ ìƒ˜í”Œë§ ì œí•œ í™•ì¸
    sample_limit = task_sample_limits.get(task, default_sample_limit)

    # ìƒ˜í”Œë§ ì¡°ê±´ í™•ì¸
    if task_counts.get(task, 0) < sample_limit and input_output_pair not in train_pairs:
        if task not in task_sampled_data:
            task_sampled_data[task] = []
        task_sampled_data[task].append(row)
        task_counts[task] += 1

    # ëª¨ë“  taskì˜ ìƒ˜í”Œë§ì´ ì™„ë£Œë˜ë©´ ì¢…ë£Œ
    if all(task_counts.get(task, 0) >= task_sample_limits.get(task, default_sample_limit) for task in task_sample_limits):
        break

# 6. ê²°ê³¼ ì¶œë ¥ ë° í™•ì¸
for task, samples in task_sampled_data.items():
    print(f"Task: {task}, Sampled: {len(samples)}")
    for sample in samples[:5]:  # ì²« 5ê°œ ìƒ˜í”Œë§Œ ì¶œë ¥
        print(sample)

# 7. í•„ìš” ì‹œ ìƒ˜í”Œë§ëœ ë°ì´í„° ì €ìž¥
# pandasë¥¼ ì‚¬ìš©í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì˜ ë”•ì…”ë„ˆë¦¬ë¥¼ ë°ì´í„°í”„ë ˆìž„ìœ¼ë¡œ ë³€í™˜ í›„ Datasetìœ¼ë¡œ ë³€í™˜
sampled_dataset = DatasetDict({
    task: Dataset.from_pandas(pd.DataFrame(samples)) for task, samples in task_sampled_data.items()
})

# ì €ìž¥ (í•„ìš”ì‹œ ì£¼ì„ ì œê±°)
# sampled_dataset.save_to_disk("sampled_test_dataset")
```

```python
import json
import re
from tqdm import tqdm
from transformers import TextStreamer

FastLanguageModel.for_inference(model)

output_file = "inference_results_0114.jsonl"

# 1. `tqdm`ì„ ì‚¬ìš©í•˜ì—¬ ì§„í–‰ ìƒíƒœ í‘œì‹œ
total_samples = sum(len(samples) for samples in sampled_dataset.values())
with open(output_file, 'w') as f_out:
    with tqdm(total=total_samples, desc="Processing samples", unit="sample") as pbar:
        for task, sample in sampled_dataset.items():
            for samp in sample:
                # ë°ì´í„°ì…‹ ì´ë¦„ì— ë”°ë¥¸ context length ì„¤ì •
                dataset_name = samp.get("dataset", "")
                max_new_tokens = 8192 if dataset_name in ["pubmed", "icliniq"] else 2048

                # ìž…ë ¥ í† í° ìƒì„±
                inputs = tokenizer(
                    [
                        alpaca_prompt.format(
                            samp['instruction'],  # instruction
                            samp['input'],  # input
                            "",  # output
                        )
                    ], return_tensors="pt"
                ).to("cuda")

                # TextStreamer ì„¤ì •
                text_streamer = TextStreamer(tokenizer)
                
                # ëª¨ë¸ ìƒì„± ë° ì¶œë ¥
                output_tensor = model.generate(
                    **inputs, 
                    max_new_tokens=max_new_tokens
                )
                model_output = tokenizer.decode(output_tensor[0], skip_special_tokens=True)

                # `### Response:` ë’¤ì˜ í…ìŠ¤íŠ¸ ì¶”ì¶œ
                response_text = None
                response_match = re.search(r"### Response:\s*(.+)", model_output, re.DOTALL)
                if response_match:
                    response_text = response_match.group(1).strip()
                else:
                    response_text = "No valid response found"

                # ëª¨ë¸ ì¶œë ¥ ê²°ê³¼ë¥¼ sampì— ì¶”ê°€
                samp['model_output'] = response_text

                # JSONL í˜•ì‹ìœ¼ë¡œ ì €ìž¥
                f_out.write(json.dumps(samp, ensure_ascii=False) + "\n")
                
                # tqdm ì§„í–‰ ìƒíƒœ ì—…ë°ì´íŠ¸
                pbar.update(1)
```
